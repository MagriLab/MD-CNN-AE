```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
with h5py.File('data/raw_pressure.h5','r') as hf:
    print(hf.keys())
    dt = np.squeeze(hf.get('dt'))
    esp = np.array(hf.get('esp'))
    nt_per_test = np.array(hf.get('nt_per_test').astype('int'))
    static_p = np.squeeze(hf.get('static_p'))
    x = np.array(hf.get('x'))
    y = np.array(hf.get('y'))
```

```python
nt_per_test
```

```python
m = np.mean(esp[:2732,:,:],axis=0)
```

```python
x_plt = np.concatenate([x,x[:,[0]]],axis=1)
y_plt = np.concatenate([y,y[:,[0]]],axis=1)
m_plt = np.concatenate([m,m[:,[0]]],axis=1)
```

```python
m_center = np.reshape([np.mean(m_plt[0,:])]*9,(1,9))
xy_center = np.reshape([0.0]*9,(1,9))

x_plt = np.concatenate([xy_center,x_plt],axis=0)
y_plt = np.concatenate([xy_center,y_plt],axis=0)
m_plt = np.concatenate([m_center,m_plt],axis=0)
```

```python
plt.figure()
plt.contourf(x_plt[:,:],y_plt[:,:],m_plt[:,:],cmap='jet',levels=100)
plt.colorbar()
```
